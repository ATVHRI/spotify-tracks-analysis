```python
# ============================================================
# Spotify Tracks Dataset — Initial Exploration & Cleaning
# ============================================================

import pandas as pd
import numpy as np

# ── Load ────────────────────────────────────────────────────
df = pd.read_csv('spotify_tracks.csv')


# ── Structure ───────────────────────────────────────────────
print("Shape:", df.shape)
print("\nFirst 5 rows:")
print(df.head())
print("\nData types and missing values:")
print(df.info())


# ── Numerical Summary ───────────────────────────────────────
print("\nDescriptive statistics:")
print(df.describe())


# ── Categorical Exploration ─────────────────────────────────
print("\nTop genres:")
print(df['track_genre'].value_counts().head(10))

print("\nTop artists:")
print(df['artists'].value_counts().head(10))

print("\nExplicit content breakdown:")
print(df['explicit'].value_counts())


# ── Missing Values ──────────────────────────────────────────
print("\nMissing values per column:")
print(df.isnull().sum())


# ── Duplicates ──────────────────────────────────────────────
print(f"\nDuplicate rows: {df.duplicated().sum()}")
df = df.drop_duplicates()
print(f"Shape after dropping duplicates: {df.shape}")


# ── Data Types ──────────────────────────────────────────────
# duration is in ms — convert to seconds for readability
df['duration_sec'] = df['duration_ms'] / 1000
df = df.drop('duration_ms', axis=1)


# ── Rename for Clarity ──────────────────────────────────────
df.rename(columns={
    'track_name': 'title',
    'track_genre': 'genre',
    'artists': 'artist'
}, inplace=True)


# ── Sorting ─────────────────────────────────────────────────
print("\nTop 10 most popular tracks:")
print(df.sort_values('popularity', ascending=False)[
    ['title', 'artist', 'genre', 'popularity']
].head(10))


# ── Filtering ───────────────────────────────────────────────
# Tracks with popularity > 80
popular = df.loc[df['popularity'] > 80]
print(f"\nHighly popular tracks (>80): {len(popular)}")

# iloc example — first 5 rows, first 4 columns
print("\niLoc sample:")
print(df.iloc[0:5, 0:4])


# ── Early Observations ──────────────────────────────────────
print("\nGenre counts:")
print(df['genre'].value_counts())
```
